# Import Library

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Data Loading

```python
df = pd.read_csv('/content/all_data.csv')
```

# Exploratory Data Analysis (EDA)

```python
# Check the data
df.head(5)
```

```python
# Check the data type & data name from all_data table
df.info()
```

```python
df.describe()
```

* The year is between 2000 to 2015.
* The Life expectancy at birth (years) is between 44.3 to 81.0 with an average of 72.8.
* The GDP is between 4.4 Billion to 18 Trillion with an average of 3.8 Trillion.

```python
# Has life expectancy increased over time in the six nations?
sns.set(style="whitegrid")
plt.figure(figsize=(12, 6))
sns.lineplot(x="Year", y="Life expectancy at birth (years)", hue="Country", data=df)
plt.title("Life Expectancy Over Time")
plt.show()
```

There is an increase in life expectancy in all six nations, and it can be seen in the country of Zimbabwe that growth is very rapid.

```python
# Has GDP increased over time in the six nations?
plt.figure(figsize=(12, 6))
sns.lineplot(x="Year", y="GDP", hue="Country", data=df)
plt.title("GDP Over Time")
plt.show()
```

There is an increase in GDP in the USA and China, but in other countries growth is constant.

```python
# Is there a correlation between GDP and life expectancy of a country?
plt.figure(figsize=(10, 8))
sns.scatterplot(x="GDP", y="Life expectancy at birth (years)", hue="Country", data=df)
plt.title("GDP vs. Life Expectancy")
plt.show()
```

Outside of Zimbabwe, we can conclude that there is a positive correlation between GDP and Life Expectancy.

```python
# What is the average life expectancy in these nations?
average_life_expectancy = df.groupby("Country")["Life expectancy at birth (years)"].mean()
print("Average Life Expectancy:")
print(average_life_expectancy)
```

Germany has the highest average life expectancy among other countries, while Zimbabwe has the lowest average life expectancy among other countries.

```python
# What is the distribution of life expectancy?
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x="Life expectancy at birth (years)", bins=10, kde=True)
plt.title("Distribution of Life Expectancy")
plt.show()
```

Life expectancy at 80 years has the highest distribution compared to other years.

# Conclusion

* Increase in Life Expectancy: We observed a consistent increase in life expectancy in all six nations over the years. Notably, this trend is most pronounced in Zimbabwe, where life expectancy has experienced rapid growth.

The results largely align with our expectations, given the global trends in healthcare and economic development. The observed increase in life expectancy, positive correlation between GDP and life expectancy (except in Zimbabwe), and variations in average life expectancy reflect known patterns in public health and economics.
